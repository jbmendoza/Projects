Problem: we will have 2574 dummy variables, thus there will be very high dimendionality.

### High dimensionality

(create a dataframe of dummy variables for 'StockCode'; add CustomerID)

```python
df.head()
```

```python
# item_dummies
item_dummies = pd.get_dummies(df.StockCode)

# add CustomerID to item_dummies
item_dummies['CustomerID'] = df.CustomerID

# first 5 rows of item_dummies
item_dummies.head()
```

### Roll up the item dummies data into customer-level item data.

```python
# aggregating at customer level
item_data = item_dummies.groupby('CustomerID').sum()

# first 5 rows of item_data
item_data.head()
```

```python
# total times each item was purchased
item_data.sum()
```

### Save item_data

```python
item_data.to_csv('item_data.csv')
```

### Thresholds

(getting only the top items to reduce dimensionality)

```python
# most popular 20 items
item_data.sum().sort_values().tail(20)
```

```python
# get list of StockCodes for the 20 most popular items
top_20_items = item_data.sum().sort_values().tail(20).index

top_20_items
```

Keep only the features for those 20 items.

```python
top_20_item_data = item_data[top_20_items]

# shape
top_20_item_data.shape
```

```python
top_20_item_data.head()
```

### Save the threshold item data

```python
# Save threshold_item_data.csv
top_20_item_data.to_csv('threshold_item_data.csv')
```
